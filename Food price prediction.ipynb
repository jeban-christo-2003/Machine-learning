# Normality testing for checking skewness

```python
from scipy.stats import shapiro


for col in df.select_dtypes(include=['int64', 'float64']).columns:

    statistic, p_value = shapiro(df[col])
    
    alpha = 0.05
    if p_value > alpha:
        print(f"The distribution of '{col}' is likely normal (fail to reject H0)")
    else:
        print(f"The distribution of '{col}' is not normal (reject H0)")
```

```python
#The distribution is not a nromal distribution since it has skewness so going for hypothesis testing with non parametric testing
```

# Hypothesis Testing

```python
from scipy.stats import mannwhitneyu

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(df['checkout_price'], df['base_price'])

# Check the p-value
alpha = 0.05
if p_value < alpha:
    print("There is a statistically significant difference between the distributions (reject H0)")
else:
    print("There is no statistically significant difference between the distributions (fail to reject H0)")
```

```python
#There is significant difference though since the base price is significantly lower with checkout price 
```
